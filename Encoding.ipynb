### **Ordinal Encoding**

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python

df["Size_en"] = oe.transform(df[["Size"]])
df
```

```python
#Ordinal Encoding using map
ord_data1 = {"s":0,"m":1,"l":2,"xl":3,} 
'''
Map use karte time hum khud decide kar sakte hai ki kisko kaunsa number milega
'''
```

```python
df['Size_en_map']=df['Size'].map(ord_data1)
```

```python
df
```
