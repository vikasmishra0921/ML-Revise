### **Outlier With Z-score**

```python
dataset2 = pd.read_csv('loan_data_set.csv')
dataset.head()
```

```python
plt.figure(figsize = (15,5))
sns.boxplot(x='CoapplicantIncome', data = dataset2)
```

```python
plt.figure(figsize = (15,5))
sns.distplot(dataset2['CoapplicantIncome'])
```

```python
min_range = dataset2['CoapplicantIncome'].mean() - (3* dataset2['CoapplicantIncome'].std())
max_range = dataset2['CoapplicantIncome'].mean() + (3 * dataset2['CoapplicantIncome'].std())
min_range , max_range
```

```python
new_data =  dataset2[dataset2['CoapplicantIncome']<=max_range] 
new_data
```

```python
sns.boxplot(x = 'CoapplicantIncome',data = dataset2) #With outliers
```

```python
sns.boxplot(x = 'CoapplicantIncome',data = new_data) #less outliers
```

## **#Real z-score **

```python


z_score = (dataset2['CoapplicantIncome'] - dataset2['CoapplicantIncome'].mean()) / (dataset2['CoapplicantIncome'].std()) 
z_score
```

```python
dataset2['z_score'] = z_score
```

```python
dataset2
```

```python
dataset2[dataset2['z_score']<3]
```
